** Hard: What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...) **

```python
# index = int(sal.Email.str.find('@')) + 1
providers = dict()

for value in sal.Email:
    host = value.split('@')[1] #get email host after "@"
    
    if host not in providers:
        providers[host] = 1
    else:
        providers[host] = providers[host] + 1
        
# print(providers)

sal2 = pd.DataFrame([providers], columns=providers.keys())
sal3 = sal2.sort_values(by=0, ascending=False, axis=1)

print(sal3.columns[0:5])   
```
